```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv(r"canada_per_capita_income.txt")
df=df.rename(columns={"per capita income (US$)":"income"},inplace=False)
df
```

```python
%matplotlib inline
plt.xlabel('year')
plt.ylabel('income')
plt.scatter(df.year,df.income,color='red',marker='+')
```

```python
new_df=df.drop('income',axis='columns')
new_df
```

```python
income=df.income
income
```

```python
#create linear regression object
reg=linear_model.LinearRegression()
reg.fit(new_df,income)
```

```python
reg.predict([[2017]])
```

```python
reg.predict([[2020]])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
#y=mx+b(m= coefficient,slope or gradient,b= intercept)
828.46507522*2020+-1632210.7578554575
```

```python
%matplotlib inline
plt.xlabel('year',fontsize=20)
plt.ylabel('income',fontsize=20)
plt.scatter(df.year,df.income,color='red',marker='+')
plt.plot(df.year,reg.predict(df[['year']]),color="blue")
```
